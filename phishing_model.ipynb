```python
import pandas as pd

data = pd.read_csv("malicious_phish.csv")
print("Level 1 done !")
data.head()
```

```python

print("Total Links Count:")
print(data['type'].value_counts())
print("-" * 30)
data['url_length'] = data['url'].apply(len)
data.head()
print("Level 2 done !")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

X = data['url']
Y = data['type']

vectorizer = TfidfVectorizer()
X_numbers = vectorizer.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_numbers, Y, test_size=0.2, random_state=42)

print("Level 3 Complete! Magic is done. ✨")
print(f"Links for Training : {X_train.shape[0]}")
print(f"Linnks for Testing : {X_test.shape[0]}")
```

```python
from sklearn.linear_model import LogisticRegression

# 1. Initialize the Logistic Regression model
model = LogisticRegression(max_iter=1000)

# 2. Train the model using the training data
model.fit(X_train, Y_train)

print("Level 4 Complete: Model Training Successful. ✨")
print("The model has learned the patterns to classify safe and phishing links.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# 1. Model ni exam lo: Ene e links aapo je ene pehla kyarey nathi joi (X_test)
predictions = model.predict(X_test)

# 2. Score check karo: Model na javab (predictions) ne sacha javab (Y_test) sathe compare karo
accuracy = accuracy_score(Y_test, predictions)

# 3. Result Declare karo!
print(f"Level 5 Complete: Model Evaluation Successful. 🏆")
print(f"Final Model Accuracy: {accuracy * 100:.2f}%")
print("-" * 40)
print("Detailed Report (Precision & Recall):")
print(classification_report(Y_test, predictions))
```
